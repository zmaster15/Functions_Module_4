## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
wholesale_df["line_subtotal"] = wholesale_df["unit_price"] * wholesale_df["qty"]
wholesale_df["line_subtotal"] = wholesale_df["line_subtotal"].astype("float64")
wholesale_df[["unit_price", "qty", "line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
wholesale_df["total_weight"] = wholesale_df.apply(lambda x: x["unit_weight"]* x["qty"], axis=1)

# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
wholesale_df["shipping_price"] = wholesale_df.apply(lambda x: x['total_weight']*7 if x['total_weight'] > 50 else x['total_weight']*10, axis=1).astype("float64")
wholesale_df[["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
wholesale_df["line_price"] = wholesale_df.apply(lambda x: (x["shipping_price"] + x["line_subtotal"]) *1.0925, axis=1).round(2)
wholesale_df[["line_subtotal", "shipping_price", "line_price"]].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
wholesale_df["line_cost"] = (wholesale_df["unit_cost"] * wholesale_df["qty"]) + wholesale_df["shipping_price"]

#wholesale_df.drop('line_subtotal)', axis=1) | encountered an additional coloumn that I added by mistake earlier
# shipping price (assume the shipping cost is exactly what is charged to the client).
wholesale_df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
wholesale_df["line_profit"] = wholesale_df.apply(lambda x: x["line_price"] - x["line_cost"], axis=1)
wholesale_df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

top_five = [33615, 66037, 46820, 38378, 24741]

# Creating new datafame with only columns desired, that i need to perfrom calculation on 
interim_df = wholesale_df[["client_id","qty","shipping_price","line_price", "line_cost"]]

# Create an empty DataFrame to store the results
top_clients_df = pd.DataFrame(columns=['client_id', 'qty', 'shipping_price', 'line_price', 'line_cost'])


for x in top_five:
    # Filter the DataFrame for the current client ID
    client_df = interim_df[interim_df['client_id'] == x]

    # Calculate the sums for the desired columns
    total_qty = client_df['qty'].sum()
    total_shipping_price = client_df['shipping_price'].sum()
    total_line_price = client_df['line_price'].sum()
    total_line_cost = client_df['line_cost'].sum()
    
    # Append the results to the results DataFrame
    top_clients_df.loc[len(top_clients_df)]= [int(x), int(total_qty), total_shipping_price, total_line_price, total_line_cost]
    

top_clients_df['client_id'] = top_clients_df["client_id"].astype('int32')
top_clients_df['qty'] = top_clients_df["qty"].astype('int32')
print(top_clients_df.head(5))
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

top_clients_df['line_profit'] = top_clients_df['line_price'] - top_clients_df['line_cost']
top_clients_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_clients_organized_df = top_clients_df.rename(columns={"client_id":"Client ID", 
                                                          "qty":"Units", 
                                                          "shipping_price": "Shipping (millions)", 
                                                          "line_price": "Total Revenue (millions)",
                                                          "line_cost": "Total Cost (millions)",
                                                          "line_profit": "Total Profit (millions)"})

# Define the money columns. 
# money_columns = ["Total Revenue (millions)", "Total Cost (millions)", "Total Profit (millions)"]
# I couldnt figure out how to isolating these columns 



# Define a function that converts a dollar amount to millions.
def millions(x):
    x = x/1000000
    return x

# Apply the currency_format_millions function to only the money columns. 
top_clients_organized_df["Shipping (millions)"] = top_clients_organized_df["Shipping (millions)"].apply(millions).round(2)
top_clients_organized_df["Total Revenue (millions)"] = top_clients_organized_df["Total Revenue (millions)"].apply(millions).round(2)
top_clients_organized_df["Total Cost (millions)"] = top_clients_organized_df["Total Cost (millions)"].apply(millions).round(2)
top_clients_organized_df["Total Profit (millions)"] = top_clients_organized_df["Total Profit (millions)"].apply(millions).round(2)


#top_clients_organized_df = top_clients_organized_df[["Total Revenue (millions)", "Total Cost (millions)", "Total Profit (millions)"]].millions()

# Rename the columns to reflect the change in the money format. 
top_clients_organized_df = top_clients_organized_df.astype({"Shipping (millions)": float, 
                                                            "Total Revenue (millions)": float, 
                                                            "Total Cost (millions)": float,
                                                            "Total Profit (millions)": float},  
                                                            errors='raise')

top_clients_organized_df.head()
```
